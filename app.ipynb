```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
file_path = ''
file_name = 'train.csv'
```

```python
data = pd.read_csv(file_path + file_name)
```

```python
data.describe()
```

```python
data.head()
```

```python
y = data.SalePrice
```

```python
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = data[features]
```

```python
X.describe()
```

```python
X.head()
```

```python
model = DecisionTreeRegressor(random_state=0)
model.fit(X, y)
```

```python
predictions = model.predict(X)
predictions
```

# Model validation

Note: This is a "In-Sample" Score -> use train_test_split

```python
mean_absolute_error(y, predictions)
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)
```

```python
model = DecisionTreeRegressor(random_state=0)
model.fit(train_X, train_y)
```

```python
predictions = model.predict(val_X)
```

```python
mean_absolute_error(val_y, predictions)
```

# Improvement with max_leaf_nodes

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    predictions = model.predict(val_X)
    mae = mean_absolute_error(val_y, predictions)
    return mae
```

```python
mae = {}
for max_leaf_nodes in range(5, 100, 5):
    mae[max_leaf_nodes] = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print(f'Max Leaf Nodes: {max_leaf_nodes} \t \t MAE: {mae[max_leaf_nodes]}')
```

# Improvment with RandomForest

```python
model = RandomForestRegressor(random_state=1)
model.fit(train_X, train_y)
```

```python
predictions = model.predict(val_X)
```

```python
mean_absolute_error(val_y, predictions)
```
